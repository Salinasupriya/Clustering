```python
#Renaming columns
data1=data.rename({'Unnamed: 0':'State'},axis=1)
data1.head()
```

```python
# Normalization function 
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
# Normalized data frame (considering the numerical part of data)
data_norm = norm_func(data1.iloc[:,1:])
data_norm.head()
```

```python
# creating dendrogram

plt.figure(figsize=(12,6))  
dendrogram = sch.dendrogram(sch.linkage(data_norm,method='complete'))
plt.show()
```

```python
# create clusters as 3
hc = AgglomerativeClustering(n_clusters=3, affinity = 'euclidean', linkage = 'complete')
hc
```

```python
y_hc = hc.fit_predict(data_norm)
```

```python
Clusters=pd.DataFrame(y_hc,columns=['Clusters'])
data1['h_clusterid'] = Clusters
```

```python
#No.of clusters formed

data1['h_clusterid'].value_counts()
```

```python
#grouping coustmers 
data1.groupby('h_clusterid').agg(['mean']).reset_index()
```

```python
#plotting the clusters based on the balance
plt.figure(figsize=(10, 7))  
plt.scatter(data1['h_clusterid'],data1['Murder'], c=hc.labels_)
plt.show()
```
